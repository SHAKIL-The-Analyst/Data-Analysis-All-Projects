__Data Cleaning__

```python
df.isnull().sum()
```

```python
sns.heatmap(df.isnull())
```

```python
Per_null= df.isnull().sum()*100/len(df)
Per_null
```

```python
df.drop('Cabin', axis= 1, inplace= True)
```

```python
df['Embarked'].mode()
```

```python
df['Embarked'].fillna('S', inplace= True)
```

```python
df['Age'].fillna(df['Age'].mean(), inplace= True)
```

```python
df.isnull().sum()
```

```python
df['Sex'].unique()
```

```python
df['Gender']= df['Sex'].map({'male':1, 'female': 0})
```

```python
df.head(1)
```

```python
x=df['Sex'].map({'male':1, 'female': 0})
df.insert(5,'Gener_New', x)
```

```python
df.head()
```

```python
df['Embarked'] = df['Embarked'].astype('category')
```

```python
pd.get_dummies(df, columns=['Embarked'], dtype=int, drop_first= True)
```

__Exploratory Data Analysis__

__How Many People Survived And How Many Died?__

```python
df['Survived'].value_counts()
```

```python
ax=sns.countplot(data= df, x='Survived', palette="Set1", saturation=1)
for container in ax.containers:
    ax.bar_label(container)
```

__How Many Passengers Were In First Class, Second Class, and Third Class?__

```python
df['Pclass'].value_counts()
```

```python
ax=sns.countplot(data=df, x='Pclass', palette="viridis", saturation=1.5)
for container in ax.containers:
    ax.bar_label(container)
```

__How Has Better Chance of Survival Male or Female?__

```python
df['Sex'].value_counts()
```

```python
ax=sns.countplot(data= df, x= 'Sex', palette='Set1', saturation=1)
for container in ax.containers:
    ax.bar_label(container)
```
