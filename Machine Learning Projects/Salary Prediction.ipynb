#### Import the libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
%matplotlib inline
```

#### import dataset & exract independent and dependent variable

```python
salary_data=pd.read_csv("F:/Skill Development- You Have No Access/Machine Learning/Salary_Data.csv")
```

```python
salary_data.head()
```

```python
x=salary_data['YearsExperience']
y=salary_data['Salary']
```

```python
x=np.array(x)
x=x.reshape(-1,1)
x
```

```python
y=np.array(y)
y=y.reshape(-1,1)
y
```

#### Visualizing the dataset

```python
sns.distplot(salary_data['YearsExperience'], kde=True, bins=10,color="b",)
```

```python
plt.figure(figsize=(10,5))
sns.countplot(y='YearsExperience', data=salary_data)
```

```python
plt.figure(figsize=(11,6))
sns.barplot(x='YearsExperience', y='Salary', data=salary_data)
```

```python
sns.heatmap(salary_data.corr(), annot=True)
```

#### Spliting the dataset into training set and test set

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(x,y, test_size=1/3, random_state=0)
```

#### Create simple linear regression to the training set

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train, y_train)
```

#### Predicting the test set result

```python
y_pred=lr.predict(x_test)
y_pred
```

```python
lr.intercept_
```

#### visualizing the train set result

```python
lr.coef_
```

```python
plt.scatter(x_train, y_train, color='b')
plt.plot(x_train, lr.predict(x_train), color='r')
plt.title('Salary ~ Experience (Train set)')
plt.xlabel('Years of experience')
plt.ylabel('Salary')
```

#### visualizing the test set result

```python
plt.scatter(x_test, y_test, color='b')
plt.plot(x_train, lr.predict(x_train), color='r')
plt.title('Salary ~ Experience (test set)')
plt.xlabel('Years of experience')
plt.ylabel('Salary')
```

#### Finding the Residuals

```python
from sklearn import metrics
print('MAE: ', metrics.mean_absolute_error(y_test, y_pred))
print('MSE: ', metrics.mean_squared_error(y_test,y_pred))
print('RMSE: ', np.sqrt(metrics.mean_absolute_error(y_test, y_pred)))
```
